# Potencia irradiada por una carga puntual acelerada
En este notebook ilustramos la potencia irradiada por unidad de angulo sólido por una carga puntual acelerada en movimiento rectilineo. Usamos las fórmulas no relativistas y relativistas obtenidas a partir de los potenciales de Lienard-Wietchert. Recuerden que cada celda a continuación se debe ejecutar apretando SHIFT+ENTER:

```python
# Importamos NumPy, Matplotlib, y las funciones especiales de SciPy
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm, colors
from scipy.special import *
```

```python
# Definimos valores adimensionales para las cantidades físicas
q = 1  # carga
c = 1  # velocidad de la luz
a = .3 # aceleración de la carga
```

## Caso no relativista

```python
theta = np.linspace(0,np.pi,100)      # linspace genera un vector con 100 valores equiespaciados entre 0 y pi
phi   = np.linspace(0,np.pi,100)      # phi toma valores entre 0 y pi (graficamos media sección)
theta, phi = np.meshgrid(theta, phi)  # meshgrid genera una malla 2D con todos los valores de theta y phi

# Potencia irradiada por unidad de ángulo sólido
P = q**2/(4*np.pi*c**3)*a**2*np.sin(theta)**2

# Calculamos las coordenadas cartesianas (x,y,z) de todos los puntos en la malla
x = P*np.sin(theta)*np.cos(phi)
y = P*np.sin(theta)*np.sin(phi)
z = P*np.cos(theta)

# Coloreamos la superficie de la esfera con el valor normalizado de la parte real del armónico esférico
fig = plt.figure(figsize=(10,6))
ax  = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z, rstride=1,cstride=1, facecolors=cm.seismic((P-np.min(P))/(np.max(P)-np.min(P))))
plt.ylim(-np.max(y),np.max(y))
plt.tight_layout()
plt.show()
```

## Caso relativista

```python
# Definimos un valor para beta (prueben cambiar este valor):
beta = .5

# Potencia irradiada por unidad de ángulo sólido
P = q**2/(4*np.pi*c**3)*a**2*np.sin(theta)**2/(1-beta*np.cos(theta))**5

# Calculamos las coordenadas cartesianas (x,y,z) de todos los puntos en la malla
x = P*np.sin(theta)*np.cos(phi)
y = P*np.sin(theta)*np.sin(phi)
z = P*np.cos(theta)

# Coloreamos la superficie de la esfera con el valor normalizado de la parte real del armónico esférico
fig = plt.figure(figsize=(10,6))
ax  = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z, rstride=1,cstride=1, facecolors=cm.seismic((P-np.min(P))/(np.max(P)-np.min(P))))
plt.ylim(-np.max(y),np.max(y))
plt.tight_layout()
plt.show()
```
