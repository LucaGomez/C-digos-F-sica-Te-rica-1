# Coeficientes de Fresnel
Graficamos la amplitud relativa y la fase de la onda transmitida y de la onda reflejada en los coeficientes de Fresnel. Usamos la raiz cuadrada de "numpy.lib.scimath" para obtener números complejos cuando la variable es negativa. Recuerden que cada celda a continuación debe ejecutarse en orden, apretando SHIFT+ENTER:

```python
# Importamos NumPy y Matplotlib
import numpy as np
import matplotlib.pyplot as plt
from numpy.lib.scimath import sqrt
```

```python
# Definimos valores para n y n' (asumimos mu = mu' = 1); prueben cambiar estos valores
n  = 1.5  # índice de refracción n  en el medio de la onda incidente
n2 = 1.   # índice de refracción n' en el medio de la onda transmitida

# Generamos un vector para el ángulo incidente
i = np.linspace(0,np.pi/2,1000)
```

## Caso transverso-eléctrico
Graficamos la amplitud relativa y la fase de la onda transmitida y de la onda reflejada en los coeficientes de Fresnel para el caso TE. Recordar que n > n' si queremos ver reflexión total interna. En ese caso, notar también la diferencia entre la amplitud de la onda transmitida, y la potencia transmitida.

```python
# Calculamos las amplitudes reflejadas (E2''/E0) y transmitidas (E2'/E0)
E2pp = (n*np.cos(i)-sqrt(n2**2-n**2*np.sin(i)**2))/(n*np.cos(i)+sqrt(n2**2-n**2*np.sin(i)**2))
E2p  = 2*n*np.cos(i)                              /(n*np.cos(i)+sqrt(n2**2-n**2*np.sin(i)**2))

# Calculamos la potencia reflejada y transmitida, la última requiere considerar el cambio de impedancia en los medios (n'/n) y la dirección de propagación (cos(t)/cos(i))
R2 = np.abs(E2pp)**2
T2 = np.abs(E2p)**2*(n2/n)*np.real(sqrt(n2**2-n**2*np.sin(i)**2))/np.cos(i)
```

```python
# Graficamos la amplitud de las ondas
plt.figure(figsize=(7,4))
plt.plot(i*180/np.pi,np.abs(E2pp),label='Reflejada')
plt.plot(i*180/np.pi,np.abs(E2p) ,label='Transmitida')
plt.xlabel(r'$i$',fontsize=16)
plt.ylabel(r'Amplitud',fontsize=16)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Graficamos la potencia
plt.figure(figsize=(7,4))
plt.plot(i*180/np.pi,R2,label='Reflejada')
plt.plot(i*180/np.pi,T2,label='Transmitida')
plt.xlabel(r'$i$',fontsize=16)
plt.ylabel(r'Potencia',fontsize=16)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Graficamos la fase de las ondas
plt.figure(figsize=(7,4))
plt.plot(i*180/np.pi,np.angle(E2pp)*180/np.pi,label='Reflejada')
plt.plot(i*180/np.pi,np.angle(E2p) *180/np.pi,label='Transmitida')
plt.xlabel(r'$i$',fontsize=16)
plt.ylabel(r'Fase',fontsize=16)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```

## Caso transverso-magnético
Graficamos la amplitud relativa y la fase de la onda transmitida y de la onda reflejada en los coeficientes de Fresnel para el caso TM. Recordar que n > n' si queremos ver reflexión total interna. En este caso también podemos ver el efecto del ángulo de Brewster en la reflexión.

```python
# Calculamos las amplitudes reflejadas (E1''/E0) y transmitidas (E1'/E0)
E1pp = (n2**2*np.cos(i)-n*sqrt(n2**2-n**2*np.sin(i)**2))/(n2**2*np.cos(i)+n*sqrt(n2**2-n**2*np.sin(i)**2))
E1p  = 2*n*n2*np.cos(i)/(n2**2*np.cos(i)+n*sqrt(n2**2-n**2*np.sin(i)**2))

# Calculamos la potencia reflejada y transmitida, la última requiere considerar el cambio de impedancia en los medios (n'/n) y la dirección de propagación (cos(t)/cos(i))
R1 = np.abs(E1pp)**2
T1 = np.abs(E1p)**2*(n2/n)*np.real(sqrt(n2**2-n**2*np.sin(i)**2))/np.cos(i)
```

```python
# Graficamos la amplitud de las ondas
plt.figure(figsize=(7,4))
plt.plot(i*180/np.pi,np.abs(E1pp),label='Reflejada')
plt.plot(i*180/np.pi,np.abs(E1p) ,label='Transmitida')
plt.xlabel(r'$i$',fontsize=16)
plt.ylabel(r'Amplitud',fontsize=16)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Graficamos la potencia
plt.figure(figsize=(7,4))
plt.plot(i*180/np.pi,R1,label='Reflejada')
plt.plot(i*180/np.pi,T1,label='Transmitida')
plt.xlabel(r'$i$',fontsize=16)
plt.ylabel(r'Potencia',fontsize=16)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.plot(i*180/np.pi,np.angle(E1pp)*180/np.pi,label='Reflejada')
plt.plot(i*180/np.pi,np.angle(E1p) *180/np.pi,label='Transmitida')
plt.xlabel(r'$i$',fontsize=16)
plt.ylabel(r'Fase',fontsize=16)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```
